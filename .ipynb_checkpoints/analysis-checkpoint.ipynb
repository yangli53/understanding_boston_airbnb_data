## Data Preparation ##

First, I need to clean the data by removing unnecessary columns, filling missing data, changing erroneous datatype etc.

### Removing unnecessary columns ###

The columns I will remove include:

- `columns lacking variability` All rows of these columns have the same value which is obviously not necessary.
- `columns with more than half missing values` I checked these columns and most of them are okay to remove except for square_foot. I think square foot may contribute to price but with very limited data this column is useless.
- `columns with url` These links are not useful for analysis.
- `columns with text information` These information may be useful but I want to focus on numeric variables and categorical variables.
- `columns not related to price/rating or have repeated information` 

### Extracting data ###

Some important variables have symbols in it, such as price with a "$" symbol. I need to remove those symbols and change the data type to float. 

### Filling missing data ###

- Some hosts may prefer not to provide name, location and neighbourhood and I will replace the missing data with "Unknown".
- Some new listings may not have host response time, host response rate and review scores yet. I will fill the mode for missing values to consider them as the average of others.
- If security deposit or cleaning fee is not listed, it may indicate that the host does not require it. I will replace the missing values with 0.
- Three listings have missing bathrooms and all of them are private rooms of houses. I guess these are shared bathrooms so I will replace them with 0.
- Three listings have missing bedrooms and two of them are entire apartments. I guess they are studios so I will replace them with 0. The other one is a private room of a house and I will replace it with 1.

### Other issues ###

- Some property types can be categorized as Apartment, House or Hotel. For example, Condominium, Serviced apartment and Loft are apartments.
- The values in host_since are dates which are not easy to evaluate. I will convert them to active days a host has been on Airbnb.
- The datatype of id is int which is not necessary. I will set id as index.


I will create a function called **clean_data( )** to implement the above operations. 

```python
def clean_data(df):
    """
    INPUT
    df - a dataframe containing original airbnb listings data in Boston
    
    OUTPUT
    df - a clean dataframe without unnecessary columns:
    1. Remove columns lacking variability.
    2. Remove columns with more than half missing values.
    3. Remove all url columns. 
    4. Remove all text columns.
    5. Remove columns which are not useful or have repeated information. 
    6. Remove % and $ from certain columns and change their data type to float - host_response_rate, price, 
       security_deposit, cleaning_fee, extra_people.
    7. Fill "Unknown" for missing values in name, host_location and host_neighbourhood.
    8. Fill the mode for missing values in host_response_time, host_response_rate, review_scores_rating, 
       review_scores_accuracy, review_scores_cleanliness, review_scores_checkin, review_scores_communication, 
       review_scores_location, review_scores_value and reviews_per_month.
    9. Fill 0 for missing values in security_deposit, cleaning_fee.
    10. Fill 0 for missing values in bathrooms.
    11. Fill 0 for missing values in bedrooms if room_type is Entire home/apt; fill 1 if room_type is Private room.
    12. Replace property_type with those are apartments, houses or hotels.
    13. Create a new column host_active_days and drop the original column.
    14. Set id as index.
    
    """
    # Remove columns lacking variability
    non_var_col = [col for col in df.columns if len(df[col].value_counts()) == 1]
    df = df.drop(non_var_col, axis=1)
    
    # Remove columns with more than half missing values
    half_nan_col = df.columns[df.isnull().mean() > 0.5]
    df = df.drop(half_nan_col, axis=1)
    
    # Remove all url columns which are not related to price
    url_col = [col for col in df.columns if '_url' in col]
    df = df.drop(url_col, axis=1)
    
    # Remove all text columns which are not related to price
    text_col = ['summary', 'space', 'description', 'neighborhood_overview', 'notes', 'transit', 'access', 
                'interaction', 'house_rules', 'host_about', 'amenities']
    df = df.drop(text_col, axis=1)
         
    # Remove columns which are not useful or have repeated information
    remove_col = ['last_scraped', 'calendar_last_scraped', 'host_id', 'host_name', 'host_verifications', 'street', 
                  'neighbourhood', 'city', 'state', 'zipcode', 'market', 'smart_location', 'latitude', 'longitude', 
                  'minimum_minimum_nights', 'maximum_minimum_nights', 'minimum_maximum_nights', 
                  'maximum_maximum_nights', 'minimum_nights_avg_ntm', 'maximum_nights_avg_ntm', 'first_review', 
                  'last_review', 'number_of_reviews_ltm', 'calculated_host_listings_count', 
                  'calculated_host_listings_count_entire_homes', 'calculated_host_listings_count_private_rooms', 
                  'calculated_host_listings_count_shared_rooms']
    df = df.drop(remove_col, axis=1)
    
    # Remove % and $ from certain columns and change their data type to float
    str_col = ['host_response_rate', 'price', 'security_deposit', 'cleaning_fee', 'extra_people']
    for col in str_col:
        df[col] = df[col].str.strip('%$').str.replace(',', '').astype(float)
    
    # Fill "Unknown" for missing values in name, host_location, host_neighbourhood
    for col in ['name', 'host_location', 'host_neighbourhood']:
        df[col] = df[col].fillna('Unknown')
    
    # Fill the mode for missing values in certain columns
    mode_col = ['host_response_time', 'host_response_rate', 'review_scores_rating', 'review_scores_accuracy', 
                'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 
                'review_scores_location', 'review_scores_value', 'reviews_per_month']
    fill_mode = lambda col: col.fillna(col.mode()[0])
    df[mode_col] = df[mode_col].apply(fill_mode, axis=0)
    
    # Fill 0 for missing values in security deposit, cleaning fee
    fee_col = ['security_deposit', 'cleaning_fee']
    df[fee_col] = df[fee_col].fillna(0)
    
    # Fill 0 for missing values in bathrooms
    df['bathrooms'] = df['bathrooms'].fillna(0)
    
    # Fill 0 for missing values in bedrooms if the room type is entire home
    # fill 1 if the property type is private room
    df['bedrooms'] = np.where(df['room_type'] == 'Entire home/apt', df['bedrooms'].fillna(0), df['bedrooms'])
    df['bedrooms'] = np.where(df['room_type'] == 'Private room', df['bedrooms'].fillna(1), df['bedrooms'])
    
    # Replace property_type with those are apartments or houses
    to_replace = {'Serviced apartment': 'Apartment', 'Loft': 'Apartment', 'Condominium': 'Apartment', 
                  'Guest suite': 'Apartment', 'Villa': 'House', 'Bungalow': 'House', 'Cottage': 'House', 
                  'Earth house': 'House', 'Guesthouse': 'House', 'Boutique hotel': 'Hotel', 'Hostel': 'Hotel', 
                  'Resort': 'Hotel', 'Houseboat': 'Boat'}

    df['property_type'] = df['property_type'].replace(to_replace)
    
    # Create a new column host_active_days and drop the original column
    df['host_since'] = pd.to_datetime(df['host_since']) 
    df['host_active_days'] = (dt.datetime(2019, 9, 22) - df['host_since']).astype('timedelta64[D]')
    df = df.drop('host_since', axis=1)
    
    # Set id as index
    df.set_index('id', inplace=True)
   
    return df
```

For popularity, there is no variable to directly show how popular a property is in the original dataset. I will create a new column called *occupancy_rate_30* to see the occupancy rate in 30 days. I will consider those with an occupancy rate at least 75% as popular homes. I will use the proportion of popular homes in each neighbourhood as their popularity.

```python
df_clean['occupancy_rate_30'] = (1 - df_clean['availability_30']/30).round(2)
df_clean['popular'] = np.where(df_clean['occupancy_rate_30'] >= 0.75, 't', 'f')
```

It looks like South Boston Waterfront is the most expensive neighbourhood with an average price of \\$312.75. To check if there is any outlier, I will create a new column called *price_per_guest*. If *extra_people* is \\$0, *price_per_guest* = *price* / *accommodates*; if *extra_people* is higher than \\$0, *price_per_guest* = (*price* + *extra_people* x (*accommodates* - *guests_included*)) / *accommodates*.

```python
# If extra_people is $0
df_clean['price_per_guest'] = round(df_clean['price']/df_clean['accommodates'], 2)
# If extra_people is high than $0
new_value = (df_clean['price'] + df_clean['extra_people']*(df_clean['accommodates'] - df_clean['guests_included']))/df_clean['accommodates']
df_clean['price_per_guest'] = np.where(df_clean['extra_people'] != 0, round(new_value, 2), df_clean['price_per_guest'])
df_clean.head()
```

Next, I will use heatmap to see if multicollinearity is present.

```python
def corr_heatmap(df, figsize=(10, 8)):
    """
    Return a heatmap showing correlations between features of the dataframe. Figure size is adjustable.
    """
    plt.figure(figsize=figsize)
    corr = df.corr()
    mask = np.zeros_like(corr)
    mask[np.triu_indices_from(mask)] = True
    sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='YlGnBu');
```

I will create dummies for categorical variables.

```python
def create_dummy(df):
    """
    INPUT:
    df - a dataframe holding all the variables of interest
    
    OUTPUT:
    df - a new dataframe that has the following characteristics:
         1. create dummies for categorical variables
         2. removes all the original columns in catigorical columns
         3. Use a prefix of the column name with an "_" for separating 
    """
    cat_cols = df.select_dtypes(include='object').columns
    for col in cat_cols:
        dummy_col = pd.get_dummies(df[col], prefix=col, prefix_sep='_', drop_first=True)
        df = pd.concat([df.drop(col, axis=1), dummy_col], axis=1)

    return df
```

I decide to use XGBoost regression model. XGBoost is a gradient boosting algorithm providing faster speed and better performance. It also has an addition of regularization to avoid overfitting.

```python
def xgb_model(df, response_col, test_size=.3, rand_state=42):
    """
    INPUT:
    df - a preprocessed dataframe holding the numerical variables and dummied categorical variables of interest
    response_col - name of the response column 
    test_size - a float indicating what proportion of data should be in the test dataset
    rand_state - an int that is provided as the random state for splitting the data into training and test 
    
    OUTPUT:
    test_score - a float reflecting r2 score on the test data
    train_score - a float reflecting r2 score on the train data
    
    This function should:
    1. Split the data into an X matrix and a response vector y
    2. Create training and test sets of data
    3. Instantiate a LinearRegression model with normalized data
    4. Fit the model to the training data
    5. Predict the response for the training data and the test data
    6. Obtain an rsquared value for both the training and test data
    """
    # Split the data into an X matrix and a response vector y
    X = df.drop(response_col, axis=1)
    y = df[response_col]
    
    # Scaling
    scaler = StandardScaler()
    X = pd.DataFrame(scaler.fit_transform(X), columns=list(X.columns))

    # Create training and test sets of data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=rand_state)
    
    # Instantiate a LinearRegression model with normalized data
    xgb_reg = xgb.XGBRegressor(objective ='reg:squarederror')
    
    # Fit the model to the training data
    xgb_reg.fit(X_train, y_train)
    
    # Predict the response for the test data and the train data
    y_test_pred = xgb_reg.predict(X_test)
    y_train_pred = xgb_reg.predict(X_train)
    
    # Obtain an rsquared value for both the test data and the train data
    test_score = r2_score(y_test, y_test_pred)
    train_score = r2_score(y_train, y_train_pred)
    
    # Obtain an MSE value for both the test data and the train data
    test_mse = mean_squared_error(y_test, y_test_pred)
    train_mse = mean_squared_error(y_train, y_train_pred)

    return test_score, train_score, test_mse, train_mse, xgb_reg, X_train, X_test, y_train, y_test
```

```python
test_score, train_score, test_mse, train_mse, xgb_reg, X_train, X_test, y_train, y_test = xgb_model(df_property, 'price')

print(f'\nTest data r2: {round(test_score, 4)} \nTrain data r2: {round(train_score, 4)}\n')
print(f'Test data MSE: {round(test_mse, 4)} \nTrain data MSE: {round(train_mse, 4)}\n')
```
